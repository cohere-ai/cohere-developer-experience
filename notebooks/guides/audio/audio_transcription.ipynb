# Audio transcription

## Overview
There are many scenarios where accurate transcription of an audio clip is very important. Examples include voice dictation for efficient note-taking and meeting intelligence for automated meeting summaries. The Cohere Transcribe model offers state-of-the-art accuracy and high efficiency in transcription use cases like these. This notebook shows how to send an audio file to the audio transcription API and read back the resulting transcript.

## Contents
- Install the Cohere Python client.
- Prepare the audio input (sample WAV).
- Call the Audio Transcriptions endpoint.
- Review the transcript.

## Setup

```python
%pip install -U cohere
```

```python
import os

import cohere

co = cohere.ClientV2(os.getenv("COHERE_API_KEY"))
```

## Audio input

The Audio Transcriptions endpoint accepts FLAC, MP3, MPEG, MPGA, OGG, and WAV files of 25MB or less. For this notebook,
you can use the following WAV example: [transcribe-model-sample-derrida-mashup.wav](data/transcribe-model-sample-derrida-mashup.wav)

```python
AUDIO_PATH = "./path/to/transcribe-model-sample-derrida-mashup.wav"
```

## Transcribe
The request is a multipart form with `model`, `language`, and `file`. Optional `temperature` (0–1) controls randomness. The Audio Transcriptions endpoint accepts FLAC, MP3, MPEG, MPGA, OGG, and WAV files of 25MB or less.

The example below calls uses the Cohere SDK client to generate an audio transcription.

```python
response = co.audio.transcriptions.create(
    model="cohere-transcribe-03-2026",
    language="en",
    file=open(AUDIO_PATH, "rb"),
)
```

## Result

```python
print(response)
```

## Conclusion
This notebook covered how to transcribe a local audio file with Cohere’s audio transcription API by sending a multipart request with `model`, `language`, and `file` and reading the transcript from the response.
